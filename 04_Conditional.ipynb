## 17. Conditional Statements :

Conditional statements are used to make decisions in a program.

#### (a) if Statement :

The **if** statement executes a block of code only when the condition is True.

#### (b) if – else Statement

The **else** block executes when the **if** condition is False.

#### (c) Nested if Statement

A **nested if** means an **if** statement inside another **if** statement.

It is used when a condition depends on another condition.

##### **Important Points to Remember :**

+ Python uses indentation instead of braces {}

+ Indentation is usually 4 spaces

+ Conditions return either True or False

+ Comparison operators (>, <, ==, !=, >=, <=) are commonly used

#### Some Examples :


```python
if 1 > 2:
    print('1 is less than 2')
```

```python
if 1 > 2:
    print('block - line 1')
    print('block - line 2')
    print('block - line 3')
print('next line')
```

```python
if 1 < 2:
    print('block - line 1')
    print('block - line 2')
    print('block - line 3')
print('next line')
```

```python
if 1 > 2:
    print('1 is less than 2')
else:
    print('1 is not less than 2')
```

```python
account_enabled = True
balance = 7000
withdraw = 700

if account_enabled and withdraw <= balance:
    print('Authorized')
else:
    print('Not Authorized')
```

```python
account_enabled = False
balance = 7000
withdraw = 700

if account_enabled and withdraw <= balance:
    print('Authorized')
else:
    if not account_enabled:
        print('account disabled')
    else:
        print('insufficient funds')
```

###  elif Statement in Python

What we saw in the previous grade example,  there's a lot of nesting and also very hard for humans to read.  But as we know that readability is really 

important in our code.  So, instead of this nested structure, Python provides an **elif** clause.

- Equivalent to a nested **if-else** .

- Do not require this double intendation .

- Easier to read .



```python
marks = 78

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
```

```python
num = 10  

if num > 0:
    print("The number is positive")
elif num < 0:
    print("The number is negative")
else:
    print("The number is zero")
```

## 18. Ternary Conditional Operator

The ternary conditional operator is a short, one-line alternative to if–else statements.

It helps write simple conditions in a compact form.


#### - Syntax = value_if_true   if   condition   else   value_if_false

         - If condition is TRUE → returns value_if_true

         - If condition is FALSE → returns value_if_false

```python
ask_price = 100
if ask_price <= 100:
    volume = 50
else:
    volume = 80
print(volume)
```

- So this is the normal way of doing the if-else statement . But if we want to make understand as per the human readability we can do by using Ternary conditional operator. So this is how we will do it : 

```python
ask_price = 100

volume = 50 if ask_price > 50 else 80

print(volume)
```

- So now this is easily understandable for the humans to read.

```python
a = 34

b = 234

max = a if a > b else b

print(max)
```
