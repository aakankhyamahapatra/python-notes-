## 19. Lists 

+ List is a container type that is it contain some elements.

- Elements are ordered sequentially and are indexed.

- Lists are mutable that is we can add, remove and replace elements.

- Lists have unbounded growth.

- It is enclosed by square bracket

len()	 ----    length of list

max()	 ----    largest value

min()	 ----        smallest value

sum()	 ----        total

sorted() ----	         sort list


```python
l = [10, 20, 30, 40, 50]
```

```python
type(l)
```

```python
l[0]
```

```python
l[4]
```

- Python also supports negative indexes as well . But it takes from -1, since -0 doesn't exist .

```python
l = [10, 20, 30, 40, 50]
```

```python
l[-3]
```

```python
l[-5]
```

- There is also a function in python called list, which will create a list . Let's see...

```python
l = list()
```

```python
l, type(l)
```

- In list we can also change the elements

```python
l = [10, 20, 30, 40, 50]
```

```python
l[3] = 7.28
l
```

so here the 3rd index element is changed to 7.28 . This goes well with negative elements as well . 

```python
l[2] = -6.75
l
```

```python
l[4] = 'Hello'
l
```

## 20. Tuple

Tuples are very similar to list . Like they are also sequential and is a container type . 

But one main difference is that, Tuples are **immutable** that is 
- once a tuple is created, **we cannot add or remove elements** & **cannot replace elements** .

- So, tuples can be created using literals enclosing a round bracket . 

- But sometimes we don't often need the () instead we can use comma.

like t = (7, 4, 2)  {This is considered as a tuple}

and t = 7, 4, 2 {is also considered as a tuple as some specific places}

- Just like the list, tuples can also contain any object like,

(1, [true, false], (3, 4))

so here, true and false is the List 

and 3 and 4 are the tuples

and the whole is also a tuple
